```python
import pandas as pd
df = pd.read_csv('sales_data.csv')
df.head()
```

```python
df.info()
```

```python
df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

revenue_by_product = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
revenue_by_product.plot(kind='bar', color='skyblue')
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Total Revenue')
plt.show()
```

```python
region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)
print("Region-wise Revenue:")
print(region_revenue)
```

```python
rep_performance = df.groupby('SalesRep')['Revenue'].sum().sort_values(ascending=False)
print("Sales Rep Performance:")
print(rep_performance)
```

```python
top_products = df.groupby('Product')['Revenue'].sum().sort_values(ascending=False)
print("Top Products by Revenue:")
print(top_products)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
top_products.plot(kind='bar', color='teal')
plt.title('Revenue by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('revenue_by_product.png')
plt.show()
```

```python
region_revenue = df.groupby('Region')['Revenue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8,5))
region_revenue.plot(kind='bar', color='coral')
plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Revenue')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('revenue_by_region.png')
plt.show()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df['Month'] = df['OrderDate'].dt.month_name()

monthly_revenue = df.groupby('Month')['Revenue'].sum()

month_order = ['January','February','March','April','May','June',
                'July','August','September','October','November','December']
monthly_revenue = monthly_revenue.reindex(month_order)

plt.figure(figsize=(10,5))
monthly_revenue.plot(kind='line', marker='o', color='purple')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('monthly_trend.png')
plt.show()
```

## Key Insights
- Amit Sharma is the top performing sales rep with highest revenue
- Laptops and Monitors are the highest revenue-generating products
- North region shows the strongest sales performance
- Revenue peaks during specific months, useful for inventory planning

```python
category_revenue = df.groupby('Category')['Revenue'].sum().sort_values(ascending=False)
print("Revenue by Category:")
print(category_revenue)

plt.figure(figsize=(7,5))
category_revenue.plot(kind='pie', autopct='%1.1f%%')
plt.title('Revenue Share by Category')
plt.ylabel('')
plt.tight_layout()
plt.savefig('category_share.png')
plt.show()
```
